# Linear Regression Implementation with Scikit-Learn

## 1. Imports and initialization

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
```

```python
from linear_regression import LinearRegression as LinearRegressionFromScratch
```

```python
# Generate synthetic dataset
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1) # y = 4 + 3X + Gaussian noise
```

```python
# Visualize the dataset
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Synthetic Linear Dataset")
plt.show()
```

## 2. Linear Regression from Scratch

```python
# Train and predict with implementation from scratch
model_scratch = LinearRegressionFromScratch(learning_rate=0.1, n_iters=1000)
model_scratch.fit(X, y)
y_pred_scratch = model_scratch.predict(X)
```

```python
# Mean Squared Error for immplementation from scratch
mse_scratch = mean_squared_error(y, y_pred_scratch)
print(f"Mean Squared Error (Scratch): {mse_scratch:.4f}")
```

## 3. Linear Regression with Scikit-Learn

```python
model_sklearn = LinearRegression()
model_sklearn.fit(X, y)
y_pred_sklearn = model_sklearn.predict(X)
```

```python
# Mean Squared Error for sklearn implementation
mse_sklearn = mean_squared_error(y, y_pred_sklearn)
print(f"Mean Squared Error (Sklearn): {mse_sklearn:.4f}")
```

```python
print("Scratch — w:", model_scratch.w, " b:", model_scratch.b)
print("Sklearn — w:", model_sklearn.coef_, " b:", model_sklearn.intercept_)
```

The values match exactly up to the 8th decimal place. The LinearRegression class from scratch is correct.

Our implementation from scratch using gradient descent converges to exactly the same solution as sklearn, which uses the Normal Equation.
